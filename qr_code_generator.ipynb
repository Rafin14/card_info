```python
import qrcode

url = "https://rafin14.github.io/card_info/"

qr = qrcode.QRCode(
    version=None,
    error_correction=qrcode.constants.ERROR_CORRECT_M,
    box_size=10,
    border=1
)

qr.add_data(url)
qr.make(fit=True)

img = qr.make_image(
    fill_color="black",
    back_color="white"
)

img.save("qr_code.png")
img

img = img.convert("RGB")  # required for PDF export
img.save("qr_code.pdf")
```
